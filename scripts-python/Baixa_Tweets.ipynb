```python
listasemente = ["helicoptero tiro",
          "helicoptero disparo"
          "helicoptero favela",
          "caveirao aereo",
          "sapao policia",
          "sapao tiro",
          "aguia policia",
          "aguia tiro",
          "aeronave favela",
          "aeronave disparo",
          "aeronave tiro"]
```

```python
# Retorna o dia que o script est√° sendo rodado 
agora = str(datetime.date.today()).replace(" ", "").replace(":", "-").replace(".", "-")
```

```python
# Define a fun√ß√£o da busca
def consulta(busca_termo):
    c = twint.Config()
    c.Search = busca_termo
    c.Lang = "pt"
    c.Pandas = True
    c.Hide_output = True
    c.Store_csv = True
    nomearquivo = busca_termo.replace(" ", "_")
    c.Output = agora + "/" + nomearquivo + "-" + ".csv"
    twint.run.Search(c)
```

```python
# Cria diret√≥rio com o dia de hoje e faz a raspagem com barra de progresso e salva os CSV
os.mkdir(agora)
pbar = ProgressBar()
for termo in pbar(listasemente):
    consulta(termo)
```

```python
# Mostra os arquivos que ser√£o importados
# TROCAR A DATA PARA A DATA DA COLETA
all_files = glob.glob(os.path.join("2020-06-20", "*.csv"))
# all_files = glob.glob(os.path.join(agora, "*.csv"))
names = [os.path.basename(x) for x in all_files]
names
```

```python
# Converte em um dataframe
df = pd.DataFrame()

for file_, name in zip(all_files, names):
    file_df = pd.read_csv(file_, encoding='utf-8', engine='python')
    file_df['file_name'] = name
    df = df.append(file_df)

# df.columns.tolist()
```
